```python
def graham_scan(points):
    def cross_product(o, a, b):
        # Calculate the cross product of vectors OA and OB.
        return (a[0] - o[0]) * (b[1] - o[1]) - (a[1] - o[1]) * (b[0] - o[0])

    # Find the point with the lowest y-coordinate (and leftmost if tied) as the starting point.
    start = min(points, key=lambda point: (point[1], point[0]))

    # Sort the points based on their polar angles from the starting point.
    sorted_points = sorted(points, key=lambda point: (math.atan2(point[1] - start[1], point[0] - start[0]), point))

    # Initialize the convex hull with the first two sorted points.
    convex_hull = [start, sorted_points[0]]

    for point in sorted_points[1:]:
        while len(convex_hull) >= 2 and cross_product(convex_hull[-2], convex_hull[-1], point) <= 0:
            # Remove the last point from the convex hull if it forms a right turn with the current point.
            convex_hull.pop()
        convex_hull.append(point)

    return convex_hull

# Example usage:
import math

points = [(0, 3), (1, 1), (2, 2), (4, 4), (0, 0), (1, 2), (3, 1), (3, 3)]
convex_hull = graham_scan(arr)
print("Convex Hull Points:")
for point in convex_hull:
    print(point)
```

```python
def partition(arr, begin, end):
    pivot = arr[end]
    i = begin - 1

    for j in range(begin, end):
        if arr[j] < pivot:
            i = i + 1
            arr[i], arr[j] = arr[j], arr[i]
    arr[i+1], arr[end] = arr[end], arr[i+1]
    return i+1

def quick_sort(arr, begin, end):
    if (begin < end):
        indx = partition(arr, begin, end)
        quick_sort(arr, begin, indx-1)
        quick_sort(arr, indx+1, end)
```
